```python
import pandas as pd
from pathlib import Path
import os
```

```python
def get_dat_data(fp):
    """读取 .dat 数据

    Args:
        fp (str or Path): 文件路径名
    """

    if not isinstance(fp, Path):
        fp = Path(fp)
    data = pd.read_csv(fp, sep='::', header=None, engine='python')
    return data
```

```python
root = Path('./ml-1m/')
```

```python
os.listdir(root)
```

```python
movies = get_dat_data(root/r'movies.dat')    # MovieID::Title::Genres
ratings = get_dat_data(root/r'ratings.dat')  # UserID::MovieID::Rating::Timestamp
users = get_dat_data(root/r'users.dat')      # UserID::Gender::Age::Occupation::Zip-code
```
